```python
import os
import re
import pandas as pd
import gensim
import collections
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
```

```python
vtt_path_list = []
folders
for root, dirs, files in os.walk("drywall_repair"):
    for file in files:
        if file.endswith(".vtt"):
             vtt_path_list.append(os.path.join(root, file))
```

```python
stop_words = stopwords.words('english')
#vtt_path_list = [os.path.join('altshiftx','vtts','build_deck',x) for x in os.listdir('drywall_repair/vtts/build_deck')]
train_strings = []
```

```python

for i,vtt_path in enumerate(vtt_path_list):
    f = open(vtt_path,'r',encoding='UTF-8')
    lines = []
    for line in f.readlines():
        line = line.strip()
        if line == '\n':
            continue
        try:
            pd.to_datetime(line.split(' ')[0])
            continue
        except:
            pass
        lines.append(re.sub('<.*?>', '', line).strip('\n'))
    lines = list(dict.fromkeys(lines))
    full_string = ''
    for line in lines:
        full_string = full_string + line + " "
    train_strings.append(full_string)
```

```python
len(train_strings)
```

```python
i=1
for train_string in train_strings:
    print(str(i) + '--' + str(train_string.count('ceiling')))
    i+=1
```
